```python
import numpy as np
pred=np.array([1,1,0,0,1])
truth=np.array([0.30,0.7,1,0,0.5])
```

```python
def mae(truth,pred):
    total_error=0
    for yt,yp in zip(pred,truth):
        total_error+=abs(yt-yp)
    print('Total Error=',total_error) 

    mae=total_error/len(pred)
    print('MAE=',mae)
```

```python
mae(truth,pred)
```

```python
np.mean(np.abs(pred-truth))
```
