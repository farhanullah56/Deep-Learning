```python
import math
def sigmoid(x):
    return 1 / (1 + math.exp(-x))
```

```python
sigmoid(2)
```

```python
sigmoid(78)
```

```python
sigmoid(-52)
```

```python
def tenh(x):
    return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))
```

```python
tenh(34)
```

```python
tenh(23)
```

```python
tenh(7)
```

```python
tenh(-56)
```

```python
tenh(-3)
```

```python
def relu(x):
    return max(0,x)
```

```python
relu(34)
```

```python
relu(-34)
```

```python
relu(-4)
```

```python
def leaky_relu(x):
    return max(0.1*x,x)
```

```python
leaky_relu(45)
```

```python
leaky_relu(-56)
```

```python
leaky_relu(4)
```

```python
leaky_relu(-5)
```
