```python
import numpy as np
pred=np.array([1,1,0,0,1])
truth=np.array([0.30,0.7,1,0,0.5])
```

```python
def MSE(pred,truth):
    totalerror=0
    for pr,tr in zip(pred,truth):
        totalerror+=(tr-pr)**2
        mse=totalerror/len(pred)
    print('Total Error= ',totalerror)
    print('MSE= ',mse)
```

```python
MSE(pred,truth)
```

```python
ms=np.mean((truth-pred)**2)
print(ms)
```

```python
ms=np.sum((truth-pred)**2)
print(ms)
```
