```python
import numpy as np
pred=np.array([1,1,0,0,1])
truth=np.array([0.30,0.7,1,0,0.5])
def log_loss(y_true,y_pred):
    epsilon=1e-15
    new=[max(i,epsilon)for i in y_pred]
    new=[min(i,1-epsilon) for i in new]
    new=np.array(new)
    return -np.mean(y_true*np.log(new)+1-y_true*np.log(1-new))
```

```python
log_loss(pred,truth)
```
